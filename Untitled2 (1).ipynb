```python
import pandas as pd
import gradio as gr
from pycaret.regression import load_model, predict_model, get_config

# Load data and model
yield_df = pd.read_csv('yield_df.csv')
model = load_model('crop_yield_model')
input_features = get_config('X_train').columns.tolist()

# Prediction function
def predict_crop_yield(*args):
    try:
        input_dict = {feature: [arg] for feature, arg in zip(input_features, args)}
        input_df = pd.DataFrame.from_dict(input_dict)
        prediction = predict_model(model, data=input_df)
        output_col = 'Label' if 'Label' in prediction.columns else prediction.columns[-1]
        result = round(prediction[output_col][0], 2)
        return f"🌿 Predicted Crop Yield: {result} hg/ha"
    except Exception as e:
        return f"❌ Error: {str(e)}"

# Build input components
inputs_ui = []
for col in input_features:
    if col == 'Year':
        continue
    if yield_df[col].dtype == 'object':
        values = yield_df[col].dropna().unique().tolist()
        inputs_ui.append(gr.Dropdown(choices=values, label=col))
    else:
        inputs_ui.append(gr.Number(label=col))

# Create custom CSS for green & white theme
custom_css = """
body { background-color: #e6f5e6 !important; }
.gradio-container, .interface, .form, .panel { background-color: #ffffff !important; border-radius: 12px; box-shadow: 0px 0px 15px rgba(0,0,0,0.1); }
textarea, input, select { background-color: #ffffff !important; border-radius: 6px; }
"""

# Launch the interface
app = gr.Interface(
    fn=predict_crop_yield,
    inputs=inputs_ui,
    outputs="text",
    title="🌾 Crop Yield Predictor",
    description="Enter crop and climate inputs to estimate yield (in hg/ha).",
    theme="default",
    css=custom_css
)

app.launch(debug=True, share=True)
```

```python
import gradio as gr
import pandas as pd

# Load chatbot QA data
qa_df = pd.read_csv("chatbot_qa.csv")

# --- Crop Yield Predictor Function ---
def predict_yield(area, item, rainfall, pesticide, temp):
    # Dummy response (replace this with your model's prediction logic)
    return f"Predicted yield for {item} in {area} is {(float(rainfall)*0.3 + float(pesticide)*0.2 + float(temp)*0.5):.2f} hg/ha"

# --- Chatbot Response Function ---
def chatbot_response(query):
    # Search for best match
    for i, q in enumerate(qa_df["Question"]):
        if query.lower() in q.lower():
            return qa_df["Answer"].iloc[i]
    return "Sorry, I couldn't find an answer to that. Please ask another question."

# --- Crop Yield UI ---
crop_inputs = [
    gr.Dropdown(choices=["Albania", "India", "Australia"], label="Area"),
    gr.Dropdown(choices=["Maize", "Wheat", "Rice"], label="Item"),
    gr.Number(label="average_rain_fall_mm_per_year"),
    gr.Number(label="pesticides_tonnes"),
    gr.Number(label="avg_temp")
]

crop_output = gr.Textbox(label="Predicted Yield (hg/ha)")
crop_interface = gr.Interface(fn=predict_yield, inputs=crop_inputs, outputs=crop_output, title="Crop Yield Predictor")

# --- Chatbot UI ---
chatbot_input = gr.Textbox(label="Ask a question", placeholder="e.g., What fertilizer is good for wheat?")
chatbot_output = gr.Textbox(label="Response")
chatbot_interface = gr.Interface(fn=chatbot_response, inputs=chatbot_input, outputs=chatbot_output, title="AgriBot - Farmer Assistant")

# --- Combine into Tabs ---
app = gr.TabbedInterface([crop_interface, chatbot_interface], tab_names=["Crop Predictor", "Chatbot"])

# Launch
app.launch()
```
